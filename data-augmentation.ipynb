```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms


image_path = r'/kaggle/input/medical-image/Testing/glioma/Te-glTr_0000.jpg'
image = Image.open(image_path)


rotate = transforms.RandomRotation(30)
hflip = transforms.RandomHorizontalFlip()
vflip = transforms.RandomVerticalFlip()
resized_crop = transforms.RandomResizedCrop(224, scale=(0.8, 1.0))
resize = transforms.Resize((224, 224))
to_tensor = transforms.ToTensor()
normalize = transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])


rotated_image = rotate(image)
hflip_image = hflip(rotated_image)
vflip_image = vflip(hflip_image)
resized_cropped_image = resized_crop(vflip_image)
resized_image = resize(resized_cropped_image)
tensor_image = to_tensor(resized_image)
normalized_image = normalize(tensor_image)

fig, axes = plt.subplots(1, 5, figsize=(15, 5))

axes[0].imshow(rotated_image)
axes[0].set_title("Step 1: Rotated")
axes[0].axis('off')

axes[1].imshow(hflip_image)
axes[1].set_title("Step 2: Horizontal Flip")
axes[1].axis('off')


axes[2].imshow(vflip_image)
axes[2].set_title("Step 3: Vertical Flip")
axes[2].axis('off')


axes[3].imshow(resized_cropped_image)
axes[3].set_title("Step 4: Resized Crop")
axes[3].axis('off')


axes[4].imshow(resized_image)
axes[4].set_title("Step 5: Resized to 224x224")
axes[4].axis('off')

plt.tight_layout()
plt.show()

print("Tensor Image Shape after Normalization:", normalized_image.shape)
```
